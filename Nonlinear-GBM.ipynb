## Геометрическое броуновское движение

Геометрическое броуновское движение - процесс, решающий стохастическое дифференциальное уравнение $d{S_t} = S_t\mu{dt} + S_t\sigma{dW_t}$. При помощи леммы Ито можно решить это уравнение, получив $S_t = S_0{e^{(\mu - \frac{sigma^2}{2})t + \sigma{W_t}}}$. Имея решение, мы можем посэмплировать пути броуновского движения и посмотреть, как оно будет выглядеть. Мы помним, что $W_t \sim N(0, t)$.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
S_0 = 100
timesteps = 100
dt = 0.1
mu = 1
sigma = 1
num_samples = 3

plt.figure(figsize=(15, 5))

for _ in np.arange(num_samples):
    sample = []
    val = S_0
    for _ in np.arange(timesteps):
        val = val*np.exp((mu - sigma**2/2)*dt + sigma*np.random.normal(0, np.sqrt(dt)))
        sample.append(val)
    plt.plot(np.arange(timesteps)*dt, sample)
    
plt.xlabel('t')
plt.ylabel('S')
plt.show()
```
